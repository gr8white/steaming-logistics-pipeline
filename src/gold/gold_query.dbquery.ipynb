```sql
-- How many shipments are currently overdue vs. at risk?
SELECT sla_status, COUNT(*) as shipment_count
FROM rapidroute.gold.sla_monitor
GROUP BY sla_status;

-- Which carriers have the most at-risk shipments?
SELECT carrier_id, sla_status, COUNT(*) as count
FROM rapidroute.gold.sla_monitor
GROUP BY carrier_id, sla_status
ORDER BY count DESC;

-- Shipments expiring in the next hour
SELECT *
FROM rapidroute.gold.sla_monitor
WHERE hours_until_sla BETWEEN 0 AND 1
ORDER BY hours_until_sla;
```
